```python
import numpy as np

# Define the dataset
x1 = np.array([1, 2, 3, 4, 5])
x2 = np.array([2, 4, 6, 8, 10])
y = np.array([5, 8, 11, 14, 17])

# Define the learning rate and number of epochs
lr = 0.01
epochs = 1000

# Initialize the weights and bias
w = np.array([0.0, 0.0])
b = 0.0

# Define the total number of samples
n = float(len(x1))

# Perform gradient descent
for i in range(epochs):
    # Calculate the predicted y values
    y_pred = np.dot(w, np.array([x1, x2])) + b
    
    # Calculate the derivatives of the weights and bias
    d_w = (1/n) * np.dot(np.array([x1, x2]), (y_pred - y))
    d_b = (1/n) * sum(y_pred - y)
    
    # Update the weights and bias
    w = w - lr * d_w
    b = b - lr * d_b
    
    # Calculate the total loss function
    loss = (1/(2*n)) * sum((y_pred - y)**2)
    
    # Print the loss every 100 epochs
    if i % 100 == 0:
        print("Epoch {}: Loss = {}".format(i, loss))
        
# Print the final weights and bias
print("Final weights: w1 = {}, w2 = {}, b = {}".format(w[0], w[1], b))
```

```python
import numpy as np

# Define the dataset
x1 = np.array(df_scaled['X1'])
x2 = np.array(df_scaled['X2'])
y = np.array(df_scaled['Y1'])
# Normalize the data
x1 = (x1 - np.mean(x1)) / np.std(x1)
x2 = (x2 - np.mean(x2)) / np.std(x2)

# Define the learning rate and number of epochs
learning_rate = 0.0001
epochs = 1000

# Initialize the weight vector and bias
w1 = 0
w2 = 0
b = 0

# Define the sigmoid activation function
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# Define the number of samples
n = len(x1)

# Define the total loss function
def total_loss(y_hat, y):
    return -np.sum(y * np.log(y_hat) + (1 - y) * np.log(1 - y_hat)) / n

# Define the partial derivative of the loss function with respect to the weight vector and bias
def partial_derivative(y_hat, y, x):
    return np.sum((y_hat - y) * x) / n

# Train the model using gradient descent
for i in range(epochs):
    # Compute the linear equation and the predictions
    z = w1 * x1 + w2 * x2 + b
    y_hat = sigmoid(z)
    
    # Compute the gradients
    dw1 = partial_derivative(y_hat, y, x1)
    dw2 = partial_derivative(y_hat, y, x2)
    db = np.sum(y_hat - y) / n
    
    # Update the weight vector and bias
    w1 -= learning_rate * dw1
    w2 -= learning_rate * dw2
    b -= learning_rate * db
    
    # Compute the total loss
    loss = total_loss(y_hat, y)
    
    # Print the loss every 100 epochs
    if i % 100 == 0:
        print(f"Epoch {i}: Loss = {loss}")
        
# Define a function to predict the label based on the trained model
def predict(x1, x2):
    z = w1 * x1 + w2 * x2 + b
    y_hat = sigmoid(z)
    return np.round(y_hat).astype(int)

# Test the model
print("Predictions:", predict(x1, x2))
```
